> Q1
``` py
    주어진 자연수가 홀수인지 짝수인지 판별해 주는 함수(is_odd)를 작성해 보자.
```

````python
# 함수선언 def 함수(매게변수): 코드
# 함수를 사용함으로써 중복코드를 줄일수있고 코드작성을 짧게 해준다. 단 한번은 함수를 선언해야함
def is_odd(number): # 함수 해더 코드가실행되면 이것만 읽고 실행은 하지않음
    if number % 2 == 1:
        return True
    else:
        return False
   
    

n = 12
if is_odd(n):
    print("%d는 홀수입니다."%(n))
else:
    print("%d는 짝수입니다."%(n))
````

> Q2
``` py
    입력으로 들어오는 모든 수의 평균 값을 계산해 주는 함수를 작성해 보자. 
    (단 입력으로 들어오는 수의 개수는 정해져 있지 않다.)
    ※ 평균 값을 구할 때 len 함수를 사용해 보자.
```

````python
# 함수로 표현
# round() 함수 : 소수점 을 입력한곳까지만 출력해줌 round(변수혹은값,n)
def get_avg():
    li = []
    # 값을 얼마나 받을것인가에 대한 입력
    count = int(input("입력할 숫자의 갯수 입력: "))
    # 총합을 받을 변수
    total = 0
    # 값을 입력받아 총합을 for문으로 구함
    for i in range(count):
        li.append(int(input("값을 입력: ")))
        total += li[i]
    # 평균계산
    avg = total/len(li)
    # 출력
    return "입력한 값의 평균:%.2f"%(avg)

print(get_avg())
````

````python
# 입력받을 리스트 변수 생성
li = []
# 값을 얼마나 받을것인가에 대한 입력
count = int(input("입력할 숫자의 갯수 입력: "))
# 총합을 받을 변수
total = 0
# 값을 입력받아 총합을 for문으로 구함
for i in range(count):
    li.append(int(input("값을 입력: ")))
    total += li[i]
# 평균계산
avg = total/len(li)
# 출력
print("입력한 값의 평균:%.2f"%(avg))
````

``` py
    다음은 "test.txt"라는 파일에 "Life is too short" 문자열을 저장한 후 
    다시 그 파일을 읽어서 출력하는 프로그램이다.

    f1 = open("test.txt", 'w')
    f1.write("Life is too short")

    f2 = open("test.txt", 'r')
    print(f2.read())
    이 프로그램은 우리가 예상한 "Life is too short"라는 문장을 출력하지 않는다. 
    우리가 예상한 값을 출력할 수 있도록 프로그램을 수정해 보자.
```

````python
#open() 하면 Input/Output String 이 생성되고 open("불러올 파일과", '작업')
# f = open("text.txt",'w') -> f.write("내용입력 후 저장됨") 경로가 안적혀있이다면 지금 작업중인 경로에 새롭게 저장된다.
# f = open("text.txt",'r') -> 파일을 읽어옴

# 만약 외부에서 무언가를 한다면 항상 예외처리라는것을 해줘야한다 밑은 파이썬 자바는 try catch를 사용

try:
    f1 = open("test.txt",'w')
    f1.write("Life is to short") # I/O가 open되면 바로 close코딩해두고 그사이에 작업을한다
except Exception as ex:
    print(ex)
finally:
    f1.close() # 열었으면 닫아야한다.
f2 = open("test.txt",'r')
print(f2.read())
f2.close()
````

> Q6
``` py
    사용자의 입력을 파일(test.txt)에 저장하는 프로그램을 작성해 보자. 
    (단 프로그램을 다시 실행하더라도 기존에 작성한 내용을 유지하고 새로 입력한 내용을 추가해야 한다.)
```

````python
dt = input("메시지를 입력해 주세요: ")

# 기존 파일의 추가 메시지 저장
with open("test.txt",'a', encoding='utf-8') as fw: # open과 close을 같이해준다.
    fw.write('\n'+dt)

with open("test.txt", 'r',encoding='utf-8') as fr:
    print(fr.read())
````

> Q7
``` py
    다음과 같은 내용을 지닌 파일 test.txt가 있다. 
    이 파일의 내용 중 "java"라는 문자열을 "python"으로 바꾸어서 저장해 보자.

    Life is too short
    you need java
    ※ replace 함수를 사용해 보자.
```

````python
with open("test.txt",'r', encoding='utf-8') as fr:
    data = fr.read().replace("java","python")

with open("test.txt",'w', encoding='utf-8') as fw:
    fw.write(data)
````
